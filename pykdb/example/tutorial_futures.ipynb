```python
from pykdb import Futures
from datetime import datetime

start = datetime(2016, 1, 4)
end = datetime(2016, 1, 10)

obj = Futures()
```

```python
obj.symbols[0:10]
```

```python
obj.names['F101-1609']
```

```python
obj.contracts['F101-1609']
```

```python
# 指定銘柄の時系列データ取得
# freq 頻度を指定('1d', '4h', '1h', '30m', '15m', '5m')
df = obj.price(date_from=start, date_to=end, symbol='F101-1609', freq='1d')
df.head(10)
```

```python
# 指定銘柄の時系列データ取得
# freq 頻度を指定('1d', '4h', '1h', '30m', '15m', '5m')
start = datetime(2016, 8, 12)
end = datetime(2016, 8, 13)
df = obj.price(date_from=start, date_to=end, symbol='F101-1609', freq='1h')
df.head(20)
```

```python
# 全銘柄の時系列データ取得
# session セッションを指定(None=全部, 'e'=夜間) 省略時は全部
df = obj.price_all(date_from=start, date_to=end, session='e')
df.head(10)
```
