```python
from pykdb import Statistics
from datetime import datetime

start = datetime(2016, 1, 4)
end = datetime(2016, 1, 10)

f = Statistics()
```

```python
f.symbols[0:10]
```

```python
f.names['T1']
```

```python
# 指定銘柄の時系列データ取得
# freq 頻度を指定('1d')
df = f.price(date_from=start, date_to=end, symbol='T1', freq='1d')
df.head(10)
```

```python
# 全銘柄の時系列データ取得
# session セッションを指定(None=全部) 省略時は全部
df = f.price_all(date_from=start, date_to=end)
df.head(10)
```
